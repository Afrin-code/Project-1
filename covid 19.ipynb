# use machine learning algorithm

```python
data=pd.read_csv("Coid19India.csv")
data
```

```python
data=data.iloc[:,1:]
data
```

```python
Day=data[data['Current Status']=='Hospitalized'].groupby(['Month','Day'])[['Num Cases']].sum()
Day
```

```python
x=np.arange(len(Day))
x
```

```python
len(Day)
```

```python
y=Day.values
y
```

```python
plt.scatter(x,y)
plt.show()
```

```python
x=x.reshape(-1,1)
x
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x,y)
```

```python
yp=regressor.predict(x)
yp
```

```python
plt.scatter(x,y)
plt.plot(x,yp)
plt.show()
```

```python
regressor.score(x,y)*100
```
